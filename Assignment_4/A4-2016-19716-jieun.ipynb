### Q1: 긍정적인 리뷰를 선별하고 random 하게 리뷰 1000개를 선택한 후 형용사의 wordcloud를 그리시오

````python
# filtering을 위해 txt 파일을 csv 파일로 변환
import pandas as pd
data = pd.read_csv('naver_ratings.txt', delimiter = '\t') # 구분자가 tab
data.columns
````

````python
data.head()
````

````python
# 긍정적인 data만 골라내기
positive = data[data.label == 1].document
positive.tail()
````

````python
positive.to_csv('positive.txt', header=False, index=False, sep='\t', mode='a')
````

````python
with open('positive.txt', 'r') as f:
    comments = f.read().splitlines()
len(comments)
````

````python
import numpy as np
chosen_comments = np.random.choice(comments,1000)
````

````python
from konlpy.tag import Komoran

tagger = Komoran()
tagged_comments = [tagger.pos(chosen_comment) for chosen_comment in chosen_comments]
# while chosen_comment == chosen_comments[10]:
#     tagged_comments
````

````python
adjective_list = []
for chosen_comment in tagged_comments:
    for word, tag in chosen_comment:
        if tag == "VA":
            word += "다"
            adjective_list.append(word)
            
# adjective_list
````

````python
from collections import Counter
adjective_counts = Counter(adjective_list)
adjective_counts.most_common()
````

````python
%matplotlib inline
import matplotlib.pyplot as plt
from wordcloud import WordCloud
````

````python
cloud = WordCloud(width=900, height=600,
                 font_path="./한마음고딕.ttf",
                 background_color="white")
````

````python
cloud = cloud.fit_words(adjective_counts)
````

````python
plt.figure(figsize=(15,20))
plt.axis('off')
plt.imshow(cloud)
plt.show()
````
