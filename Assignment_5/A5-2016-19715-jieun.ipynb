## 1. 미디어다음에서 제공하는 영어사전 웹사이트에서 랜돔으로 100개의 단어를 수집하고, 단어와 단어의 의미를 보기 좋게 출력하시오.

예: http://dic.daum.net/word/view.do?wordid=ekw000078303&q=help


* 저장 형식은 자유. 단, 가능한 한 보기 좋게 내용을 편집하여 저장하자.

  예)
  
    단어 1 
      1. 뜻풀이  
      2. 뜻풀이 
    ===  
    단어 2 
      1. 뜻풀이

```python
import random

# 단어 개수 ekw000000001~ekw000388825(직접 count했습니다.)
url_list = []
for i in range(100):
    word = "ekw000" + str(random.randrange(1,38825)).zfill(6)
    url = "http://dic.daum.net/word/view.do?wordid={}".format(word)
    url_list.append(url)
```

```python
# 출력만 하는 것인지, 저장도 하는 것인지 헷갈려 둘 다 작성했습니다.

from bs4 import BeautifulSoup
import urllib.request
import json
from collections import OrderedDict

fp = open('words.txt', 'w')
for url in url_list:
    with urllib.request.urlopen(url) as url:
        doc = url.read()
        soup = BeautifulSoup(doc, "html.parser")
        word = soup.find("span", class_="txt_cleanword").text
        print(word)
        fp.write(word)
        meanings = soup.find_all("ul", class_="list_mean")[0].findChildren("li")
        for meaning in meanings:
            print(meaning.text)
            fp.write(meaning.text + "\n")
        print("\n===")
        fp.write("\n===\n")
fp.close()

```

```python
# 2. tweepy Cursor를 이용한 키워드 분석

keyword = "퀸덤"
result = []

fp = open('tweet_cursor.txt', 'w')
for tweet in tweepy.Cursor(api.search, q=keyword, since='2019-10-31', count=5).items(100):   
    print("{}\nby: {} retweets: {}".format(tweet.text, tweet.user.screen_name, tweet.retweet_count))
    print("\n=====")
    fp.write(tweet.user.screen_name + '\n')
    fp.write(tweet.text + '\n')
    fp.write(str(tweet.retweet_count) + '\n')
    fp.write("\n===\n")

fp.close()
    
```

```python
with open('tweet_cursor.txt', 'r') as fp:
    tweets = fp.read().split("===")
len(tweets)
```

```python
# -*- coding: utf-8 -*-
from konlpy.tag import Komoran
import re

data_list = [] 
for tweet in tweets:
    hangul = re.compile('[^ \u3131-\u3163\uac00-\ud7a3]+')  
    data_list.append(hangul.sub('', tweet))
# print(data_list)

tagger = Komoran()
tagged_tweets = []

for data in data_list:
    try:
        tag = tagger.pos(data)
        tagged_tweets.append(tag)
    except:
        print("error")

tagged_tweets
```

```python
noun_list = []
for tweet in tagged_tweets:
    for word, tag in tweet:
        if tag in ['NNP', 'NNG']:
            noun_list.append(word)
noun_list
            
```

```python
from collections import Counter
noun_counts = Counter(noun_list)
noun_counts.most_common()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
cloud = WordCloud(width=900, height=600,
                 font_path="./한마음고딕.ttf",
                 background_color="white")
```

```python
cloud = cloud.fit_words(noun_counts)
```
