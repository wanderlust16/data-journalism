```python
against_boys = []
with open('qd_idol.txt', 'r') as f:
    for line in f:
        if "남자" in line:
            against_boys.append(line)

f.close()

with open('qd_idol_vs_boys.txt', 'w') as fb:
    for line in against_boys:
        fb.write(line)

fb.close()

against_boys
```

```python
from ckonlpy.tag import Twitter as Okt

okt = Okt()
okt.add_dictionary('퀸덤', 'Noun')
okt.add_dictionary('아이돌', 'Noun')
okt.add_dictionary('남자아이돌', 'Noun')
okt.add_dictionary('남자아이돌', 'Noun')
okt.add_dictionary('마마무', 'Noun')
okt.add_dictionary('진취적', 'Adjective')
okt.add_dictionary('전투적', 'Adjective')
okt.add_dictionary('여자아이돌', 'Noun')
okt.add_dictionary('여자아이돌들', 'Noun')
okt.add_dictionary('에이오에이', 'Noun')
okt.add_dictionary('남자새끼', 'Noun')
okt.add_dictionary('컴백쇼', 'Noun')
okt.add_dictionary('여돌', 'Noun')
okt.add_dictionary('여돌들', 'Noun')
okt.add_dictionary('남돌', 'Noun')
okt.add_dictionary('성상품화', 'Noun')
okt.add_dictionary('걸크러시', 'Noun')
okt.add_dictionary('걸크러쉬', 'Noun')

stopwords=['거','것','보','해','말','함','그','뭐','더','이','보고','안','고','나','까지','적','개','생','하나','데','화','게','수','때','없','어서','너',
          '못', '있', '요', '줄', '여서', '이나', '임', '저', '어도', '오', '건', '아하', '면', '볼', '애', '온', '내', '기', '단']
```

```python
tagged_data = []
for line in against_boys:
    tagged_data.append(okt.pos(line))

tagged_data
```

```python
from collections import Counter

word_list = []
for wordset in tagged_data:
    for word, tag in wordset:
        if word not in stopwords and tag in ["Noun", "Adjective"]:
            word_list.append(word)
                    
word_counts = Counter(word_list)
word_counts_list = word_counts.most_common()
word_counts_list

# for (word, count) in word_counts_list:
    
# word_counts_list
# len(word_counts_list) 
```

```python
major_words = []
for word, count in word_counts_list:
    if count >= 6:
        major_words.append(word)
            
len(major_words)
```

```python
major_words_index = {word: i for i, word in enumerate(major_words)}
major_words_index # dict type
```

```python
import numpy as np # len(tagged_data)는 문장 개수
occurs = np.zeros([len(tagged_data), len(major_words)]) # 가로, 세로 0으로 채워진 행렬
np.shape(occurs) # (96 * 55의 행렬 데이터)
```

```python
for i, sent in enumerate(tagged_data): # 전체 문장들에서
    for word, tag in sent: # 각 문장의 word와 tag 중 
        if word in major_words: # word가 major_words라면
            index = major_words_index[word] # index는, 위의 major_words_index에 있는 값(0~54)
            occurs[i][index] += 1 # co-occurence가 있으면 그 자리에 1을 채움
            
# i: 0~95 (sentences의 개수), index: 0~54(major words의 개수)
# len(occurs[0]) # 55
occurs[0] # 55(열)
```

```python
co_occurs = occurs.T.dot(occurs) # 원행렬과 역행렬의 곱 
co_occurs
```

```python
import networkx as nx
graph = nx.Graph()

for i in range(len(tagged_data)): # 96개 data
    for j in range(i + 1, len(major_words)): # 55개 data
        if 6 < co_occurs[i][j] <= 10: 
            graph.add_edge(major_words[i], major_words[j], color="#d0bfff", width=1.0)
        elif 10 < co_occurs[i][j] <= 14: 
            graph.add_edge(major_words[i], major_words[j], color="#9775fa", width=5.0)
        elif co_occurs[i][j] > 14: 
            graph.add_edge(major_words[i], major_words[j], color="#6741d9", width=9.0)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
    
d = dict(graph.degree)

# node_colors = []
# print(graph.degree)
# for word, degree in graph.degree:
#     if degree <= 3:
#         node_colors.append("#d0bfff")
#     if 3 < degree < 8:
#         node_colors.append("#9775fa")
#     else:
#         node_colors.append("#6741d9")
        
edges = graph.edges()
# print(edges)
weights = [graph[u][v]['width'] for u,v in edges]
colors = [graph[u][v]['color'] for u,v in edges]

# print(graph["퀸덤"]["여자"])
plt.figure(figsize=(20, 16))
# layout = nx.spring_layout(graph, k=.5)

nx.draw_kamada_kawai(graph, width=weights,
        layout=nx.spring_layout(graph, k=0.05, iterations=30),
        node_color="#9775fa", node_size=[v * 800 for v in d.values()], with_labels=True,
        font_size=30, font_color="black", font_weight="bold", font_family="AppleGothic",
        edge_color=colors, alpha=0.8)

plt.show()
len(graph.edges)
```
