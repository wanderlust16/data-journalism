```python
import re
from konlpy.tag import Komoran



def strip_e(alist):
    RE_EMOJI = re.compile('[\U00010000-\U0010ffff]', flags=re.UNICODE)
    processed_alist =[]
    for item in alist :
        processed_item = RE_EMOJI.sub(r"", item)
        processed_alist.append(processed_item)
    return processed_alist
#[출처] koNLPy] kernal shut-down 될때!  이모티콘/기호 제외하기|작성자 데이터공방

def isHangul(text):
    encText = text
    hanCount = len(re.findall(u'[\u3130-\u318F\uAC00-\uD7A3]+', encText))
    return hanCount > 0
#https://blog.naver.com/chandong83/221142971719


for item in ["AOA", "마마무", "오마이걸", "박봄", "러블리즈", "아이들"]:
    with open("comments_data/comments_{}.txt".format(item), encoding="utf-8") as f:
            lines = f.read().splitlines()

    processed_lines = strip_e(lines)
    sentences = [line for line in processed_lines if (line != "") and (line != "=")]
    #print(sentences)

    tagger = Komoran()
    tagged_sentences = [tagger.pos(sent) for sent in sentences if isHangul(sent)]
    #tagged_sentences

    word_list = []
    for sent in tagged_sentences:    
        for word, tag in sent:
            if tag in ['NNP', 'NNG']:
                word_list.append(word)

    from collections import Counter
    word_counts = Counter(word_list)
    word_counts.most_common()

    %matplotlib inline
    import matplotlib.pyplot as plt
    from wordcloud import WordCloud

    cloud = WordCloud(width=900, height=600, 
                      font_path='08서울남산체 B.ttf', 
                      background_color='white')

    cloud = cloud.fit_words(word_counts) 

    plt.figure(figsize=(15, 20))
    plt.axis('off') 
    plt.imshow(cloud)
    save
```
