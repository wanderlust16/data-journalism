```python
import re
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import konlpy
from ckonlpy.tag import Twitter as Okt
import numpy as np

okt = Okt()
for item in ["퀸덤", "queendom", "컴백", "팬덤", "아이돌","걸그룹", "여돌","팬도라","걸크러시","걸크러쉬","여가수","팬덤",
             "오마이걸","러블리즈", "마마무", "에이오에이", "AOA", "여자 아이들", "여자아이들", "아이들", "G-IDLE","박봄", "IDLE", 
             "효정", "미미", "유아", "승희", "지호", "비니", "아린",
             "지민","설현","찬미","혜정","유나",
             "케이", "베이비소울", "미주","지수","예인", "지애","진", "jin","수정",
             "화사","솔라","휘인","문별",
             "소연","전소연","수진", "민니", "미연", "슈화", "우기",
             "너나해", "짧은치마", "짧은 치마", "너나 해", "쏘리", "sorry",
             "굿럭", "우린 결국 다시 만날 운명이었지", "데칼코마니", "i miss you",
             "Fire", "파이어", "싫다고말해", "싫다고 말해", "라타타", "라이언", "lion", "latata",
             "유앤아이", "눈코입", "you and i", "되돌릴 수 없는 돌아갈 수 없는 돌아갈 곳 없는", "2ne1", 
             "비밀정원", "데스티니", "destiny", "트와일라잇", "twilight", "게릴라",
             "moonlight", "문라이트", "식스센스", "아츄", "카메오", "까메오", "cameo",
            "엘비스", "무무","미라클", "러블리너스", "블랙잭", "네버랜드"]:
    okt.add_dictionary(item, "Noun")
stopwords=["진짜","정말","까지", "면서","보","이","저","그","게", "거", "것", "더","얘", "려고","들", "때","요","보고","좀","여기","부터","건", "어요", "왜"]

def strip_e(alist):
    RE_EMOJI = re.compile('[\U00010000-\U0010ffff]', flags=re.UNICODE)
    processed_alist =[]
    for item in alist :
        processed_item = RE_EMOJI.sub(r"", item)
        processed_alist.append(processed_item)
    return processed_alist
#[출처] koNLPy] kernal shut-down 될때!  이모티콘/기호 제외하기|작성자 데이터공방


```

```python
for artist in ["AOA","오마이걸","러블리즈","마마무","박봄", "아이들"]:
    with open("data/korean_comments/korean_comments_{}.txt".format(artist), "r", encoding="utf-8") as f:
            lines = f.read().split("\n=\n")

    processed_lines = strip_e(lines)
    sentences = [line for line in processed_lines if (line != "") and (line != "=")]
    #print(sentences)
    #print(len(sentences))

    tagged_sentences = [okt.pos(sent) for sent in sentences]
    #print(tagged_sentences)

    word_list = []
    for sent in tagged_sentences:    
        for word, tag in sent:
            if (word not in stopwords) and (tag in ["Noun"]):
                word_list.append(word)

    from collections import Counter
    word_counts = Counter(word_list)
    word_counts.most_common()

    %matplotlib inline
    import matplotlib.pyplot as plt
    from wordcloud import WordCloud

    cloud = WordCloud(width = 1000, height = 800,
                      font_path="08서울남산체 B.ttf", 
                      background_color="white")

    cloud = cloud.fit_words(word_counts) 

    plt.figure(figsize=(20, 15))
    plt.axis('off') 
    plt.imshow(cloud)
    plt.savefig("data/wordclouds/korean_comments_wordcloud_{}.png".format(artist))
```
