```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd

nltk.download("vader_lexicon")
sid = SentimentIntensityAnalyzer()


for artist in ["AOA","오마이걸","마마무","박봄","러블리즈","아이들"]:
    f_eng = open("data/english_comments/english_comments_{}.txt".format(artist), "r" , encoding="utf-8")
    f_ko_to_eng = open("data/ko_to_eng_comments/ko_to_eng_{}.txt".format(artist), "r" , encoding="utf-8")
    f_eng_text = f_eng.read()
    f_ko_to_eng_text = f_ko_to_eng.read()    
    merged_text = f_ko_to_eng_text + f_eng_text
    merged_text_list = merged_text.split("\n=\n")
    comments_and_sentiment=[]
    for text in merged_text_list:
        ss = sid.polarity_scores(text)
        """print(text)
        for k in sorted(ss):            
            print('{0}: {1}, '.format(k, ss[k]), end='')
        print("\n")"""
        if (ss["pos"] < ss["neg"]):
            comments_and_sentiment.append([text,ss["compound"], ss["neg"], ss["neu"], ss["pos"], "neg"]) #negative
        elif (ss["pos"] == ss["neg"]) : 
            comments_and_sentiment.append([text,ss["compound"], ss["neg"], ss["neu"], ss["pos"], "neu"]) #neutral
        else :
            comments_and_sentiment.append([text,ss["compound"], ss["neg"], ss["neu"], ss["pos"], "pos"]) #positive   
    df = pd.DataFrame(comments_and_sentiment,columns=["comment", "compound", "negative" , "neutral", "positive", "whether_pos_or_neg"])
    df.to_csv("data/comments_and_sentiment/sentiment_analysis_{}.txt".format(artist), encoding="utf-8")
```
